```python
import matplotlib.pyplot as plt
import pandas as pd

# Data from the tables provided
data_without_code = {
    "Models": ["Bart-large-mnli", "GPT3.5", "GPT-4", "Few-shot_all-MiniLM"],
    "Accuracy": [0.49, 0.35, 0.38, 0.57],
    "Precision": [0.57, 0.52, 0.48, 0.59],
    "Recall": [0.50, 0.37, 0.39, 0.55],
    "F1": [0.47, 0.32, 0.35, 0.56]
}

data_with_code = {
    "Models": ["Bart-large-mnli", "GPT3.5", "GPT-4", "Few-shot_all-MiniLM"],
    "Accuracy": [0.27, 0.29, 0.32, 0.47],
    "Precision": [0.31, 0.41, 0.31, 0.52],
    "Recall": [0.29, 0.32, 0.34, 0.51],
    "F1": [0.25, 0.26, 0.26, 0.48]
}

# Creating DataFrames
df_without_code = pd.DataFrame(data_without_code)
df_with_code = pd.DataFrame(data_with_code)

# Plotting the differences in performance
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(14, 10))

metrics = ['Accuracy', 'Precision', 'Recall', 'F1']
for i, metric in enumerate(metrics):
    ax = axes[i//2, i%2]
    ax.bar(df_without_code['Models'], df_without_code[metric], width=0.4, label='Without Code', align='center')
    ax.bar(df_with_code['Models'], df_with_code[metric], width=0.4, label='With Code', align='edge')
    ax.set_title(f'{metric} Comparison')
    ax.set_ylabel(metric)
    ax.set_xticklabels(df_without_code['Models'], rotation=45)
    ax.legend()

plt.tight_layout()
plt.show()
```
