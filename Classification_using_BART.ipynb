```python
from transformers import pipeline
classifier = pipeline("zero-shot-classification",
                      model="facebook/bart-large-mnli", use_cuda=True)
```

```python
classifier.model
```

```python
import pandas as pd

df = pd.read_csv('/content/test.csv', sep=',')
```

```python
df.head()
```

```python
zero = []
```

```python
from time import sleep
from tqdm import tqdm

for i in tqdm(range(len(df))):

  sequence_to_classify = str(df['Paper'][i]) + '' + str(df['keywords'][i]) # Forth experiment TDK
  candidate_labels = ['Bioinformatics', 'Environmental Sciences', 'Social Sciences', 'Statistics and Data Analysis', 'Economics']
  result = classifier(sequence_to_classify, candidate_labels)
  zero.append(result['labels'][0])
```

```python
zero
```

```python
df['Zero-shot'] = zero
df.to_csv('results_fb_TD.csv', index=False)
```
